```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

dataset = pd.read_csv('./datasets/radiomics_features.csv', sep=';', index_col=0).sample(frac=1)
dataset = dataset[glrlm_features]
dataset = dataset[dataset['diagnosis_code'] <= 3]

X = dataset.iloc[:, :-1]
Y = dataset.iloc[:, -1]
seed = 123
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
clf = svm.SVC(gamma='scale')
clf.fit(X_train, y_train)
accuracy = accuracy_score(y_test, clf.predict(X_test))
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from xgboost import XGBClassifier
import pickle

dataset = pd.read_csv('./datasets/radiomics_features.csv', sep=';', index_col=0).sample(frac=1)

dataset = dataset[glrlm_features]
print(dataset.head())
for i in range(1, 6):
    disease_norma_df = dataset[(dataset['diagnosis_code'] == 0) | (dataset['diagnosis_code'] == i)].sample(frac=1)
    X = disease_norma_df.iloc[:, :-1]
    Y = disease_norma_df.iloc[:, -1]
    
    print(X, Y)
    seed = 123
    test_size = 0.2
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
    
    model = XGBClassifier()
    model.fit(X_train, y_train)

    pickle.dump(model, open("./models/xgb/xgb_" + diseases[0] + "_" + diseases[i] + ".pickle.dat", "wb"))
    
    
    y_pred = model.predict(X_test)
    predictions = [round(value) for value in y_pred]
    accuracy = accuracy_score(y_test, predictions)
    print(diseases[i].upper() + " Accuracy: %.2f%%" % (accuracy * 100.0))
```
